```python
print(1 and 2 and 3 and 4) # 1
print(1 or 2 or 3 or 4) # 2
print(1 and 2 or 3 and 4) # 3
print((1 and 2) or (3 and 4)) # 4
print(1 or 2 and 3 or 4) # 5
print((1 or 2) and (3 or 4)) # 6
```

1. 모두 and 논리연산자를 사용했기 때문에 1, 2, 3을 차례로 평가하고 최종적으로 4와 평가되어진 값이 반환된다.
2. 모두 or 논리연산자를 사용했기 때문에 두 개의 값중 하나 이상이 True이면 True를 평가한다. 처음에 1이라는 True라는 값을 가지기 때문에 한 값 1만 반환된다.
3. 1의 값이 True인 것을 확인한 후 그 다음 값과 연산하여 2의 값이 True로 확인된다. 그 다음 or 연산자는 앞이나 뒤 값이 True이면 True를 반환한다. and가 or보다 우선순위가 높으므로 2 or 4 연산을 거쳐 2라는 값이 반환된다.
4. 괄호 안에 있는 연산을 진행하면 2 or 4이라는 최종적인 식이 나온다. 2의 값이 True임을 확인한 후에 True를 반환하고 실행을 멈춘다.
5. and 연산자를 먼저 진행하면 1 or 3 or 4가 되는데 1이 True의 값을 가지므로 1의 값을 반환한다.
6. 괄호 안에 있는 연산을 수행하면 1 and 3 이라는 식이 나오게 되고 1의 값이 True이므로 3의 값을 최종적으로 반환하게 된다.

```python
def div_path(a) : # 함수 선언
    num = 0
    #L = len(a)
    name = '' # 경로명 초기화
    
    for i in a : # 입력된 문자열의 각 문자를 읽기 위한 for문 실행
        num += 1 # 인덱스 한 칸씩 증가
        if( i == '(') : # '(' 괄호가 나올 때 까지 각 문자를 읽고 
            break # 괄호가 나오면 종료
        
    a = a[num:] # num번째 인덱스로부터 슬라이싱
    a = a.strip("')") # ') 제거
    a = a.split('/') # '/'을 기준으로 문자열을 분리
    a = a[1:] # '' 공백을 제거
    
    b = [] # 공 리스트 선언
    for i in range(len(a)-1) : # 입력한 문자열 길이 -1만큼 반복문 실행
        name = name + '/' + a[i] # 마지막 파일명을 제외하고 경로명 설정
    b.append(name) # 공리스트였던 b에 추가
    b.append(a[len(a)-1]) # 파일명 추가
    return b # 나뉘어진 경로와 파일명 리턴
    
a = input() # 경로 입력
print( div_path(a) )
```

사용자로부터 경로를 입력받는다. 받은 경로를 문자열로 설정하여 각 인덱스에 저장된 문자를 이용하여 필요없는 문자들은 과감히 슬라이싱한다.
if문을 이용하여 '('라는 괄호가 나오면 해당 인덱스 번호를 num이라는 변수에 저장한다. '(' 전에 나왔던 문자들은 필요없으므로 제거하고 그외
불필요한 괄호와 따옴표를 strip을 이용하여 제거한다. split을 이용하여 '/'를 기준으로 문자열을 분리한다. 
그리고 공리스트를 선언하여 우리가 필요한 경로와 파일명을 대입한다. 기존 문자열의 길이만큼 파일명을 제외한 필요 경로명을 name이라는 문자형
변수에 저장한다. 문자형 변수를 append를 이용하여 추가하고 파일명 역시 삽입해준다. 최종적인 리스트 b를 리턴한다.

```python
def list_union(lista, listb) :

    for i in lista :
        if(i not in listb) : # lista의 원소가 listb에 존재하는지 판단
            listb.append(i) # listb에 삽입
    for i in range(len(listb)-1) : # listb 확인
        if(listb[i] in listb[i+1:]): # 중복된거 있는지 확인
            del listb[i] # 중복이면 중복된 인덱스 제거
            
    return sorted(listb) # 정렬한 listb를 반환

print (list_union([1,2,3], [1,2,4]))
```

두 개의 lista, listb에서 for문을 이용하여 lista의 원소가 listb에 없으면 listb에 append를 이용하여 원소를 추가한다.
listb의 원소가 중복되는 것을 방지하기 위해 삭제한다. 최종적으로 정렬된 listb를 반환한다.

```python
import requests

req = requests.get('http://cse.koreatech.ac.kr')

source = req.text
```

```python
print(source)
```

```python
source = source.strip() # 공백제거
string = '' # 빈 문자열
check = True # boolean check 초기화
l = [] # 빈 리스트

for i in range(len(source)): # 소스코드의 길이만큼 반복
    if (source[i] == '<' ): # < 가 나오면 false로 설정
        check = False
        continue
    if ( source[i] == '>' and check == False): # >가 나오면 true
        check = True
        continue
    if ( check == True ): # < > 문자를 제외한 단어들을 ch에 저장
        string += source[i]

l = string.split() # 문자열 ch를 공백을 기준으로 l이라는 리스트에 저장
print ( len(l) ) # 리스트의 길이 출력
```

strip()을 이용하여 소스코드에 공백들을 제거한다. 그리고 빈 문자열과 리스트 그리고 boolean형 check를 초기화 선언한다. 반복문을 이용하여 소스코드 전체 문자들을 확인하는 작업을 수행한다. < > 문자를 제외한 단어들을 continue를 이용하여 check가 true일 때 string에 저장한다. 저장한 문자들을 l이라는 빈 리스트에 공백을 기준으로 저장한다. 그리고 리스트의 길이를 출력한다.
